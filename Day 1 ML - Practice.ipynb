```python
#Dimensionality, Type, Average Horsepower Value, Correlated Features.
import pandas as pd
import numpy

df = pd.read_csv("mtcars.csv",  low_memory=False)
df.isna().any()
```

```python
display(df)
```

```python
df.describe()
```

```python
df[['hp',  'mpg', 'disp', 'drat']].groupby(['hp']).describe().unstack()
```

```python
df[['hp',  'mpg', 'disp',  'drat']].corr()
```

```python
# Dimentianlity

print("Dimensions of mtCar.csv: " + str(df.shape))

# Type 

print("Type of mtCar.csv: " + str(type(df)))

# Type of Columns

print(str(df['hp'].dtype))
```

```python
# Slicing Dataframe

print(df.iloc[2:6,0:4])
```

```python
# Identifying Unique Elements

print(df['mpg'].unique())
```

```python
# Average Horsepower Value

print("Average (mean) Horsepower is: " + str(df['hp'].mean()))
print("Average (median) Horsepower is: " + str(df['hp'].median()))
```

```python
# Mode

display(df.mode(axis=0))
```

```python
# Correlated  Features

import matplotlib.pyplot as plt
import seaborn as sns
correlations = df.corr()
sns.heatmap(data=correlations,square=True,cmap="bwr")

plt.yticks(rotation=0)
plt.xticks(rotation=90)
```

```python
sns.boxplot(x=df['mpg'])
```
